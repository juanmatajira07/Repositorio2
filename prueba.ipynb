```python
def funcion():    
    import pandas as pd
    import numpy as np

    # Set a random seed for reproducibility
    np.random.seed(47) # You can use any integer value here. For quiz put 42

    # Generate 25 random names for the 'nombre' column
    nombres = [f'Estudiante{i}' for i in range(10)]

    # Generate 4 columns of random grades between 0 and 5 with two decimal places
    # np.random.uniform generates floats between low (inclusive) and high (exclusive)
    parciales = np.round(np.random.uniform(0, 5, size=(10, 4)), 2);

    # Create the DataFrame
    df = pd.DataFrame({
        'nombre': nombres,
        'parcial1': parciales[:, 0],
        'parcial2': parciales[:, 1],
        'parcial3': parciales[:, 2],
        'parcial4': parciales[:, 3]
    })
    
    df['promedio']=(df['parcial1']*0.20+df['parcial2']*0.20+df['parcial3']*0.20+df['parcial4']*0.40)
    cont=0
    for i in df['promedio']:
        if i>=3:
            cont+=1 
    return f'{cont} personas lograron pasar'
```
